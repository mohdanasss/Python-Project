```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sb
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding ='unicode_escape')
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1 , inplace= True)
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df[['Age','Amount']].describe()
```

# Exploratory Data Analysis

```python
df.columns
```

```python
ax = sb.countplot(x='Gender',data = df,palette = 'magma')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gender = df.groupby(["Gender"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sb.barplot(x='Gender',y='Amount',data=sales_gender, palette='icefire')
```

From above graphs we can see that most of the buyers are females and even the puchasing power of females are greater men. 

# Age

```python
df.columns
```

```python
ax = sb.countplot(x='Age Group',data = df,hue = 'Gender',palette = 'magma')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(["Age Group"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sb.barplot(x='Age Group',y='Amount',data=sales_age,palette='icefire')
```

From above graph we can see that of the buyers are of age group between 26-35

# State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)

sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(data = sales_state, x = 'State',y= 'Orders',palette= 'magma')
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)

sb.set(rc={'figure.figsize':(15,5)})
sb.barplot(data = sales_state, x = 'State',y= 'Amount',palette='icefire')
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

# Marital Status

```python
ax = sb.countplot(data = df, x = 'Marital_Status',palette ='magma')

sb.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sb.set(rc={'figure.figsize':(6,5)})
sb.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender',palette='icefire')
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

# Occupation

```python
sb.set(rc={'figure.figsize':(20,5)})
ax = sb.countplot(data = df, x = 'Occupation',palette= 'tab10')

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

# Product Category

```python
sb.set(rc={'figure.figsize':(20,5)})
ax = sb.countplot(data = df, x = 'Product_Category',palette = 'magma')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',palette='tab10')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})
sb.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',palette = 'magma')
```

```python

fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
